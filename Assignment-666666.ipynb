```python
class Dog {
constructor(name, age, coatColor) {
this.name = name;
this.age = age;
this.coatColor = coatColor;
}

description() {
console.log(${this.name} is ${this.age} years old.)
}

get_info() {
console.log(${this.name}s coat color is ${this.coatColor})
}
}

class JackRussellTerrier extends Dog {
constructor(name, age, coatColor) {
super(name, age, coatColor);
}

breed_description() {
console.log(${this.name} is a Jack Russell Terrier)
}

energy_level() {
console.log(${this.name} is a high energy dog)
}
}

class Bulldog extends Dog {
constructor(name, age, coatColor) {
super(name, age, coatColor);
}

breed_description() {
console.log(${this.name} is a Bulldog)
}

energy_level() {
console.log(${this.name} is a low energy dog)
}
}

let bruno = new Dog('Bruno', 5, 'brown');
let buddy = new JackRussellTerrier('Buddy', 8, 'white');
let teddy = new Bulldog('Teddy', 4, 'black');

bruno.description();
bruno.get_info();

buddy.description();
buddy.get_info();
buddy.breed_description();
buddy.energy_level();

teddy.description();
teddy.get_info();
teddy.breed_description();
teddy.energy_level();
```
